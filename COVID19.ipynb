```python
def plotChartTotalDeaths(countryData):
    countryData.groupby('total_deaths').head(10).plot(x='date', y='total_deaths', figsize=(15,10))
    plt.title('Total Deaths COVID-19 '+ countryData['location'].iloc[0])
    plt.ylabel('Quantity')
    plt.xlabel('Date')
    plt.savefig('Charts/Others/TotalDeaths/%s - COVID19.png'%countryData['location'].iloc[0])
def plotChartTotalCases(countryData):
    countryData.groupby('total_cases').head(10).plot(x='date', y='total_cases', figsize=(15,10))
    plt.title('Total cases COVID-19 '+ countryData['location'].iloc[0])
    plt.ylabel('Quantity')
    plt.xlabel('Date')
    plt.savefig('Charts/Others/TotalCases/%s - COVID19.png'%countryData['location'].iloc[0])
    
def plotChartNewCases(countryData):
    countryData.groupby('new_cases').head(10).plot(x='date', y='new_cases', figsize=(15,10))
    plt.title('New Cases COVID-19'+ countryData['location'].iloc[0]+'.png')
    plt.ylabel('Quantity')
    plt.xlabel('Date')
    plt.savefig('Charts/Others/NewCases/%s - COVID19.png'%countryData['location'].iloc[0])

def plotChartNewDeaths(countryData):
    countryData.groupby('new_deaths').head(10).plot(x='date', y='new_deaths', figsize=(15,10))
    plt.title('New deaths COVID-19 ' + countryData['location'].iloc[0])
    plt.ylabel('Quantity')
    plt.xlabel('Date')
    plt.savefig('Charts/Others/NewDeaths/%s - COVID19.png'%countryData['location'].iloc[0])
    
    
```

```python
def plotDataByCountry(country):
    
    countryData = database.query('location=="%s"'%country)
    plotChartTotalDeaths(countryData)
    plotChartTotalCases(countryData)
    plotChartNewCases(countryData)
    plotChartNewDeaths(countryData)

```

## Functions
    In the next cells has some functions to manipulate the data and create folders with the country name and create the 
    chart correspondent to the country that was created a folder.


```python
''' Create a directory in charts folder with the parameter dirName
    parameters:: dirName -> str 
'''
def createDir(dirName):
    import os
    import re
    try:
        dirName = re.sub('[^A-Za-z0-9]+', '', dirName)
        os.mkdir('Charts/Others/'+dirName)
        print("Directory " , dirName ,  " Created ")
    except FileExistsError as e:
        return e
        #print("Directory " , dirName ,  " already exists")

def createDirInCountry(dirName, newDir):
    import os
    import re
    try:
        dirName = re.sub('[^A-Za-z0-9]+', '', dirName)
        os.mkdir('Charts/Others/'+dirName+'/'+newDir)
        #print("Directory inside" , dirName ,  " Created ")
    except FileExistsError as e:
        return e
        #print("Directory Inside " , dirName ,  " already exists")
```

```python
import datetime

'''
    Format the string to remove special characters.
'''
def removeSpecialChars(string):
    import re
    if("Cote d'Ivoire" in string):
        string.replace("Cote d'Ivoire", "Cote dIvoire")
    string = re.sub('[^A-Za-z0-9]+', '', string)
    return string

'''
    return a dataframe filtered by the country.
'''
def getDataByCountry(dataframe, country):
    if(country == "Cote d'Ivoire"):
        data = database.query('location=="%s"'%"Cote d'Ivoire")
    else:
        data = database.query('location=="%s"'%country)
    return data;

'''
    Create a folder for each country and inside there, a new folder, to save the chart of country  based in the new deaths. In the end of this process
    the spent time to make all charts is showed.
'''
def plotChartNewDeathsAllCountries(dataframe, countryList):
    start = datetime.datetime.now()
    print("Creating Charts to New Deaths")
    for country in countryList:
        countryData = getDataByCountry(dataframe, country)
        createDir(removeSpecialChars(country))
        
        countryNewDeaths = countryData.groupby('date').new_deaths.sum()
        countryNewDeaths.tail(7).plot(kind='bar', rot=90, figsize=(15,15))
        plt.xlabel("Date")
        plt.ylabel("Quantity")
        plt.title('New deaths COVID-19 %s' % country)
        createDirInCountry(removeSpecialChars(country), 'New Deaths')
        plt.savefig('Charts/Others/'+removeSpecialChars(country)+'/New Deaths/%s - New Deaths - COVID19.png'%removeSpecialChars(country))
    print("Time:", datetime.datetime.now() - start)
    
    
'''
    Create a folder for each country and inside there, a new folder, to save the chart of country  based in the new cases. In the end of this process
    the spent time to make all charts is showed.
'''        
def plotChartNewCasesAllCountries(dataframe, countryList):
    start = datetime.datetime.now()
    print("Creating Charts to New Cases")
    for country in countryList:
        countryData = getDataByCountry(dataframe, country)
        createDir(removeSpecialChars(country))
        countryNewCases = countryData.groupby('date').new_cases.sum()
        countryNewCases.tail(7).plot(kind='bar', rot=90, figsize=(15,15))
        plt.xlabel("Date")
        plt.ylabel("Quantity")
        plt.title('New cases COVID-19 %s' % country)
        createDirInCountry(removeSpecialChars(country), 'New Cases')
        plt.savefig('Charts/Others/'+removeSpecialChars(country)+'/New Cases/%s - New Cases - COVID19.png'%removeSpecialChars(country))
    print("Time:", datetime.datetime.now() - start)
 
'''
    Create a folder for each country and inside there, a new folder, to save the chart of country  based in the total cases. In the end of this process
    the spent time to make all charts is showed.
'''    
def plotChartTotalCasesAllCountries(dataframe, countryList):
    start = datetime.datetime.now()
    print("Creating Charts to Total Cases")
    for country in countryList:
        countryData = getDataByCountry(dataframe, country)
        createDir(removeSpecialChars(country))
        
        countryTotalCases = countryData.groupby('date').total_cases.sum()
        countryTotalCases.tail(7).plot(kind='bar', rot=90, figsize=(15,15))
        plt.xlabel("Date")
        plt.ylabel("Quantity")
        plt.title('Total Cases COVID-19 %s' % country)
        createDirInCountry(removeSpecialChars(country), 'Total Cases')
        plt.savefig('Charts/Others/'+removeSpecialChars(country)+'/Total Cases/%s - Total Cases - COVID19.png'%removeSpecialChars(country))
    print("Time:", datetime.datetime.now() - start)
        
'''
    Create a folder for each country and inside there, a new folder, to save the chart of country  based in the total deaths. In the end of this process
    the spent time to make all charts is showed.
'''    
def plotChartTotalDeathsAllCountries(dataframe, countryList):
    start = datetime.datetime.now()
    print("Creating Charts to Total Deaths...")
    for country in countryList:
        countryData = getDataByCountry(dataframe, country)
        createDir(removeSpecialChars(country))
        
        countryTotalDeaths = countryData.groupby('date').new_deaths.sum()
        countryTotalDeaths.tail(7).plot(kind='bar', rot=90, figsize=(15,15))
        plt.xlabel("Date")
        plt.ylabel("Quantity")
        plt.title('Total Deaths COVID-19 %s' % country)
        createDirInCountry(removeSpecialChars(country), 'Total Deaths')
        plt.savefig('Charts/Others/'+removeSpecialChars(country)+'/Total Deaths/%s - Total Deaths - COVID19.png'%removeSpecialChars(country))
    print("Time:", datetime.datetime.now() - start)

        
```
