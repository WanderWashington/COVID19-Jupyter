```python
def plotChartTotalDeaths(countryData):
    countryData.groupby('total_deaths').head(10).plot(x='date', y='total_deaths', figsize=(15,10))
    plt.title('Total Deaths COVID-19 '+ countryData['location'].iloc[0])
    plt.ylabel('Quantity')
    plt.xlabel('Date')
    plt.savefig('Charts/TotalDeaths/%s - COVID19.png'%countryData['location'].iloc[0])
def plotChartTotalCases(countryData):
    countryData.groupby('total_cases').head(10).plot(x='date', y='total_cases', figsize=(15,10))
    plt.title('Total deaths COVID-19 '+ countryData['location'].iloc[0])
    plt.ylabel('Quantity')
    plt.xlabel('Date')
    plt.savefig('Charts/TotalCases/%s - COVID19.png'%countryData['location'].iloc[0])
    
def plotChartNewCases(countryData):
    countryData.groupby('new_cases').head(10).plot(x='date', y='new_cases', figsize=(15,10))
    plt.title('Total deaths COVID-19'+ countryData['location'].iloc[0]+'.png')
    plt.ylabel('Quantity')
    plt.xlabel('Date')
    plt.savefig('Charts/NewCases/%s - COVID19.png'%countryData['location'].iloc[0])

def plotChartNewDeaths(countryData):
    countryData.groupby('new_deaths').head(10).plot(x='date', y='new_deaths', figsize=(15,10))
    plt.title('Total deaths COVID-19 ' + countryData['location'].iloc[0])
    plt.ylabel('Quantity')
    plt.xlabel('Date')
    plt.savefig('Charts/NewDeaths/%s - COVID19.png'%countryData['location'].iloc[0])
    
    
```

```python
def plotDataByCountry(country):
    countryData = database.query('location=="%s"'%country)
    plotChartTotalDeaths(countryData)
    plotChartTotalCases(countryData)
    plotChartNewCases(countryData)
    plotChartNewDeaths(countryData)
```

```python
plotDataByCountry('Brazil')
plotDataByCountry('Spain')
plotDataByCountry('Italy')
plotDataByCountry('United States')
```
