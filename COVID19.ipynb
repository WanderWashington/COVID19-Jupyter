```python
import pandas as pd 
import numpy as np
import csv
import json
import requests
import urllib.request as downloader
import matplotlib.pyplot as plt
%matplotlib inline
import requests

print('Downloading Dataset')

url = 'https://covid.ourworldindata.org/data/ecdc/full_data.csv'
r = requests.get(url)

with open('Data/full_data.csv', 'wb') as f:
    f.write(r.content)
    print('Done!')
    
```

```python
database = pd.read_csv('Data/full_data.csv', sep=',')
```

```python

brazilCases.sort_values(by="date")#kind='bar', figsize=(15,5))
# brazilCases.set_index("date",drop=True,inplace=True)

brazilCases.sort_values(by='date', ascending=False).head(10).plot(kind='bar', figsize=(15,10), x='date', rot=0)
plt.title('COVID-19 Brazil - 02/04/2020')
plt.ylabel('Quantity')
plt.xlabel('Date')

plt.savefig('Charts/BrazilCasesPerDay.png')
```

## Charts COVID-19 Brazil

    *New Deaths;
    *New Cases;
    *Total Cases;
    *Total Deaths;
    

```python
# total_deaths = brazilCases["total_deaths"].value_counts()
# total_deaths.groupby('total_deaths').plot(kind='barh', x='date', y='total_deaths')

brazilCases.groupby('new_deaths').head(10).plot(x='date', y='new_deaths', figsize=(15,10))
plt.title('New deaths COVID-19 - Brazil - 02/04/2020')
plt.ylabel('Quantity')
plt.xlabel('Date')
plt.savefig('Charts/New deaths COVID19Brazil02042020.png')
```

```python
# total_deaths = brazilCases["total_deaths"].value_counts()
# total_deaths.groupby('total_deaths').plot(kind='barh', x='date', y='total_deaths')

brazilCases.groupby('new_cases').head(10).plot(x='date', y='new_cases', figsize=(15,10))
plt.title('New cases COVID-19 - Brazil - 02/04/2020')
plt.ylabel('Quantity')
plt.xlabel('Date')
plt.savefig('Charts/New cases COVID19Brazil02042020.png')
```

```python
# total_deaths = brazilCases["total_deaths"].value_counts()
# total_deaths.groupby('total_deaths').plot(kind='barh', x='date', y='total_deaths')

brazilCases.groupby('total_cases').head(10).plot(x='date', y='total_cases', figsize=(15,10))
plt.title('Total cases COVID-19 - Brazil - 02/04/2020')
plt.ylabel('Quantity')
plt.xlabel('Date')
plt.savefig('Charts/Total cases COVID19Brazil02042020.png')
```

```python
# total_deaths = brazilCases["total_deaths"].value_counts()
# total_deaths.groupby('total_deaths').plot(kind='barh', x='date', y='total_deaths')

brazilCases.groupby('total_deaths').head(10).plot(x='date', y='total_deaths', figsize=(15,10))
plt.title('Total deaths COVID-19 - Brazil - 02/04/2020')
plt.ylabel('Quantity')
plt.xlabel('Date')
plt.savefig('Charts/Total deaths COVID19Brazil02042020.png')
```
